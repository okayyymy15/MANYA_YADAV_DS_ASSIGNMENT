```python
## CODE FOR WEATHER DATA STORAGE SYSTEM    (MANYA YADAV(BTECH. CSE DS) SEM-3)

n = int(input("Enter how many years: "))
years = []
for i in range(n):
    y = int(input("Enter year: "))
    years.append(y)


m = int(input("Enter how many cities: "))
cities = []
for j in range(m):
    c = input("Enter city: ")
    cities.append(c)


data = [[None for _ in range(m)] for _ in range(n)]


while True:
    print("\n1. Insert")
    print("2. Retrieve")
    print("3. Delete")
    print("4. Row-Major Access")
    print("5. Column-Major Access")
    print("6. Exit")
    choice = input("Enter choice: ")

    if choice == "1":  
        print("Available years with indexes:")
        for i in range(len(years)):
            print(i, ":", years[i])
        row = int(input("Enter index of year: "))
        print("Available cities with indexes:")
        for j in range(len(cities)):
            print(j, ":", cities[j])
        col = int(input("Enter index of city: "))

        if 0 <= row < n and 0 <= col < m:
            t = float(input("Enter temperature: "))
            data[row][col] = t
            print("Inserted successfully.")
        else:
            print("Invalid index.")

    elif choice == "2":  
        print("Available years with indexes:")
        for i in range(len(years)):
            print(i, ":", years[i])
        row = int(input("Enter index of year: "))

        print("Available cities with indexes:")
        for j in range(len(cities)):
            print(j, ":", cities[j])
        col = int(input("Enter index of city: "))

        if 0 <= row < n and 0 <= col < m:
            if data[row][col] is not None:
                print("Temperature for", cities[col], "in", years[row], ":", data[row][col])
            else:
                print("No temperature stored.")
        else:
            print("Invalid index.")

    elif choice == "3":  
        print("Available years with indexes:")
        for i in range(len(years)):
            print(i, ":", years[i])
        row = int(input("Enter index of year: "))
        print("Available cities with indexes:")
        for j in range(len(cities)):
            print(j, ":", cities[j])
        col = int(input("Enter index of city: "))

        if 0 <= row < n and 0 <= col < m:
            if data[row][col] is not None:
                data[row][col] = None
                print("Deleted successfully.")
            else:
                print("Nothing to delete.")
        else:
            print("Invalid index.")

    elif choice == "4":  
        print("Row-Major Access (row by row):")
        for row in data:
            print(row)

    elif choice == "5":  
        print("Column-Major Access (column by column):")
        for c in range(len(cities)):
            col_data = [data[r][c] for r in range(len(years))]
            print(col_data)

    elif choice == "6": 
        print("Exiting...")
        break

    else:
        print("InvalidÂ choice.")
        


```
